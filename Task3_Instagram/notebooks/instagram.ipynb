# One-Page Instagram Strategy Document for Alfido Tech

## Goal
To identify optimal Instagram posting times, content types with high engagement, and recommend strategies to increase overall engagement for Alfido Tech, leveraging insights from the provided comments data.

## Key Findings from Comments Data Analysis

### 1. Optimal Posting Times (Based on Comment Frequency)
*   **Hour of Day**: The data strongly indicates that **8 AM (08:00)** is the peak hour for comment activity. All comments in the dataset were recorded at this specific hour.
*   **Day of Week**: **Thursdays** emerged as the most active day for comments. All comments in the dataset occurred on a Thursday.

**Recommendation**: To maximize immediate comment engagement, Alfido Tech should prioritize posting content on **Thursdays at 8 AM**.

### 2. Engaging Content Types (Based on Comment Frequency)
*   **Emoji Usage**: Comments with emojis (`yes`) and without emojis (`no`) had an equal number of occurrences (3744 each). This suggests that while emojis are used, their presence doesn't exclusively drive a higher *volume* of comments within this dataset.
    *   **Recommendation**: Continue using emojis as appropriate to maintain a conversational tone, but do not solely rely on them to boost comment numbers. Focus on the content's value.

*   **Hashtag Usage**: Analysis of hashtag counts in comments revealed a clear pattern:
    *   **2 Hashtags**: Most frequent, with 2496 comments.
    *   **1 Hashtag**: Second most frequent, with 1872 comments.
    *   Other counts (0, 3, 4, 5, 6 hashtags) had significantly fewer comments (624 each).

    **Recommendation**: To optimize for comment engagement, Alfido Tech should aim to use **1 to 2 hashtags** per post. This range appears to resonate best with the audience, encouraging higher comment volumes.

## Generate Synthetic Dataset for Follower Growth Analysis

To simulate the missing data for follower growth and other engagement metrics, I will create a synthetic dataset. This dataset will include a `date` range, a `follower_count` that shows a general upward trend, and simulated `likes_per_post` and `shares_per_post`.

```python
import numpy as np

# Generate a date range for the synthetic data
dates = pd.date_range(start='2023-01-01', end='2023-03-31', freq='D')

# Simulate follower count with an upward trend and some daily fluctuation
initial_followers = 1000
growth_rate = 10 # average daily growth
fluctuation = 5 # random daily fluctuation

follower_count = []
current_followers = initial_followers
for _ in range(len(dates)):
    current_followers += np.random.normal(growth_rate, fluctuation)
    follower_count.append(max(0, int(current_followers))) # Ensure non-negative followers

# Simulate likes per post and shares per post
# These can also have trends or be influenced by follower count
likes_per_post = (np.array(follower_count) * np.random.uniform(0.05, 0.15, len(dates))).astype(int)
shares_per_post = (np.array(follower_count) * np.random.uniform(0.01, 0.03, len(dates))).astype(int)

# Create the synthetic DataFrame
df_follower_growth = pd.DataFrame({
    'date': dates,
    'follower_count': follower_count,
    'likes_per_post': likes_per_post,
    'shares_per_post': shares_per_post
})

# Display the first few rows of the synthetic dataset
display(df_follower_growth.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create plots directory if it doesn't exist
plots_dir = '/content/plots'
os.makedirs(plots_dir, exist_ok=True)

# Load dataset
df = pd.read_csv("/content/comments.csv")

# Rename column for consistency and easier access
df.rename(columns={'Hashtags used count': 'hashtag_count'}, inplace=True)

# Parse datetime using 'created Timestamp'
df['created Timestamp'] = pd.to_datetime(df['created Timestamp'], format='%d-%m-%Y %H:%M')
df['hour'] = df['created Timestamp'].dt.hour
df['day_of_week'] = df['created Timestamp'].dt.day_name()

# Engagement metrics (This section is removed as these metrics are intended for the synthetic dataset)
# if 'followers' in df.columns:
#     df['likes_per_follower'] = df['likes'] / df['followers']
#     df['comments_per_follower'] = df['comments'] / df['followers']

# --- Plot 1: Engagement by Hour ---
plt.figure(figsize=(10,6))
sns.countplot(x=df['hour'], hue=df['hour'], palette="viridis", legend=False)
plt.title("Comment Frequency by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Comments")
plt.tight_layout()
plt.savefig(os.path.join(plots_dir, "engagement_by_time.png"))
plt.close()

# --- Plot 2: Engagement by Day ---
day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
comments_by_day = df['day_of_week'].value_counts().reindex(day_order)

plt.figure(figsize=(10,6))
sns.barplot(x=comments_by_day.index, y=comments_by_day.values, hue=comments_by_day.index, palette="magma", legend=False)
plt.title("Comment Frequency by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Comments")
plt.tight_layout()
plt.savefig(os.path.join(plots_dir, "engagement_by_day.png"))
plt.close()

# --- Plot 3: Hashtag Usage ---
# Using the renamed 'hashtag_count' column
if 'hashtag_count' in df.columns:
    plt.figure(figsize=(10,6))
    sns.countplot(x=df['hashtag_count'], hue=df['hashtag_count'], palette="coolwarm", legend=False)
    plt.title("Hashtag Usage Distribution")
    plt.xlabel("Number of Hashtags")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(os.path.join(plots_dir, "hashtags_distribution.png"))
    plt.close()

print("✅ Analysis complete. Plots saved in /content/plots folder.")
```
